```julia
# using Pkg
# Pkg.add("Plots")
using Plots
```

```julia
r1 = range(0.01,1.99,length = 100)
r2 = range(2.01,6,length = 100)
r = []
append!(r,r1,r2)

#we choose M = 1

#v = constant, choose a vector of v
vin = range(0,0,length = 200)
tilde_in = vin .- r

#v = 2(r + 2Mln|r/2M - 1| + const)
vout = 2(r .+ 2*log.(abs.(r/2 .- 1)))
tilde_out = vout .- r

#for both constants i have set them to 0
```

```julia
plot()
plot!(r,tilde_in,legend = false,label = false, color = :red,
    title = "Spacetime Diagram in Modified Coordinates",xlabel = 'r',ylabel = "tÌƒ", ylims = (0,15))
plot!(r,tilde_out,linestyle = :dashdot, color = :blue,label = false)
plot!([2], seriestype = "vline", color = :black ,label = "Event Horizon", linewidth = 5)
plot!([0],seriestype = "vline",color = :green, linestyle = :dash,label = "Singularity")


#adding a number is changing the constant

for i in 1:19
    plot!(r,tilde_in .+ i, color = :red,label = false)
    plot!(r,tilde_out .+ i,linestyle = :dashdot, color = :blue,label = false)
end

plot!(r,tilde_in .+ 20,legend = true, color = :red, label = "Inwards")
plot!(r,tilde_out .+ 20,legend = true ,linestyle = :dashdot, color = :blue, label = "Outwards")
```

```julia
tin = vin .- r - 2log.(abs.(r/2 .- 1))
tout = vout .- r - 2log.(abs.(r/2 .- 1))

plot()
plot!(r,tin,legend = true,label = false, color = :red,
    title = "Spacetime Diagram in Schwarzchild Metric",xlabel = 'r',ylabel = "t", ylims = (0,15))
plot!(r,tout,linestyle = :dashdot, color = :blue, label = false)
plot!([2], seriestype = "vline", color = :black ,label = "Event Horizon", linewidth = 5)
plot!([0],seriestype = "vline",color = :green, linestyle = :dash,label = "Singularity")


for i in 1:19
    plot!(r,tin .+ i,legend = false, color = :red, label = false)
    plot!(r,tout .+ i,legend = false ,linestyle = :dashdot, color = :blue, label = false)
end


plot!(r,tin .+ 20,legend = true, color = :red, label = "Inwards")
plot!(r,tout .+ 20,legend = true ,linestyle = :dashdot, color = :blue, label = "Outwards")
```
