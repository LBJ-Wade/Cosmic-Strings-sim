```julia
# using Pkg
# Pkg.add("DifferentialEquations")
# Pkg.add("Plots")
using DifferentialEquations
using Plots
```

```julia
function 🧮cceler🧮tion(🇸🇸)
    if 🇸🇸 < 1
        🧮 = 0
    elseif  1 ≤ 🇸🇸 < 26
        🧮 = 0.2
    elseif 26 ≤ 🇸🇸 < 76
        🧮 = -0.2
    elseif 76 ≤ 🇸🇸 < 101
        🧮 = 0.2
    else
        🧮 = 0
    end
    return 🧮
end



function deriv🧮tive(y,p,🇸🇸)
    x₀,x₁,u₀,u₁  = y
    dy🇸🇸 = [u₀,u₁, sign(u₁)*p(🇸🇸)/sqrt((u₀/u₁)^2-1), p(🇸🇸)/sqrt(1- (u₁/u₀)^2)]
    return dy🇸🇸
end
```

```julia
🇸🇸ₛ = (0.0,102.0)
y = [0,0,1,0]
prob = ODEProblem(deriv🧮tive,y,🇸🇸ₛ,🧮cceler🧮tion)
sol = solve(prob, Vern9(), reltol=1e-10, abstol=1e-10,
    tstops = (1,26,76,101),saveat=(0:0.01:102))

```

```julia
t = []
x = []
vₜ = []
vₓ = []

for i ∈ 1:length(sol)
    append!(t,sol[i][1]) 
    append!(x,sol[i][2])
    append!(vₜ,sol[i][3])
    append!(vₓ,sol[i][4])
end

🇸🇸 = range(0,102,length = length(t))
```

```julia
plot(x,t,legend = false,title ="Laboratory Frame Observing Rocketman"
    ,xlabel = 'x',ylabel = 't')
```

```julia
plot(t,vₓ./vₜ,legend = false,title ="Laboratory Frame Observing Rocketman Velocity"
    ,xlabel = 't',ylabel = "dx/dt")
```

```julia
plot(🇸🇸,-vₜ.^2 .+ vₓ.^2,legend = false,title ="Length of 4-velocity",
    xlabel = "τ",ylabel = "u⋅u",ylims =(-1.00000001,-0.99999999))
```

```julia
plot(🇸🇸,t,legend = false,title="Comparing Twin Clocks",xlabel = "Rocket Frame Time (τ)",
    ylabel = "Laboratory Frame Time (t)")
```
