```julia
f(x) = 1/x^2 * ( 1/r_0^2 * (1 - 2M/r_0) - 1/x^2 * ( 1 - 2M/x)  )^(-0.5)
```

```julia
r = range(4,stop = 6,length = 1000)
r_1 = 10
M = 1
delphi = []

for i in 1:length(r)
    r_0 = r[i]
    a,b = quadgk(f,r_0,r_1,rtol = 1e-3)
    append!(delphi,a)
end

for i in 1:length(r)
    delp = delphi[i]
    if delp < Ï€/2
        println(r[i])
        print(delp)

        break
        
    end
    
end
```
