```julia
# using Pkg
# Pkg.add("DifferentialEquations")
# Pkg.add("Plots")
using DifferentialEquations
using Plots
```

```julia
function derivative_p(y,p,t)
    üçµ,r,œï,u‚Çú,u·µ£,u·µ© = y 
    m = 1
    dydœÑ = [u‚Çú,u·µ£,u·µ©, -2m/r^2 *(1 - 2m/r)^(-1) * u‚Çú * u·µ£ , 
        (1 - 2m/r)*( -m/r^2 * u‚Çú^2 + 
            m/r^2* (1 - 2m/r)^(-2)* u·µ£^2 + r* u·µ©^2), -2/r * u·µ© * u·µ£ ]
    return dydœÑ
end
```

```julia
#for a photon
R,C,œâ,m = 10, 0.5, 0, 1
Œª = (0.0, 10.0)
y = [0,R,0,sqrt( ( ( R^2* C^2 - œâ)*R) /(R - 2m)), 0, C ]


prob = ODEProblem(derivative_p,y,Œª)
sol = solve(prob,Tsit5(),reltol= 1e-10,
    abstol = 1e-10,saveat = (0:0.01:10))
```

```julia
xs = []
ys = []
```

```julia
t = []
r = []
œï = []
u‚Çú = []
u·µ£ = []
u·µ© = []

for i ‚àà 1:length(sol)
    append!(t,sol[i][1])
    append!(r,sol[i][2])
    append!(œï,sol[i][3])
    append!(u‚Çú,sol[i][4])
    append!(u·µ£,sol[i][5])
    append!(u·µ©,sol[i][6])
end

x = r .* cos.(œï)
y = r .* sin.(œï)

append!(xs,[x])
append!(ys,[y])


plot(x,y,legend = true,label = "C = 0.5",
    title = "Path taken by light",xlabel = 'x',ylabel = 'y')
```
